```python
def create_magic_square():
    # Initialize 3x3 grid with zeros
    magic_square = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]

    # Starting position for 1
    row = 0
    col = 1

    # Number to be placed
    num = 1

    while num <= 9:
        magic_square[row][col] = num

        # Calculate the next position
        next_row = (row - 1) % 3
        next_col = (col + 1) % 3

        # If the cell is already filled, move vertically down
        if magic_square[next_row][next_col] != 0:
            row = (row + 1) % 3
        else:
            row, col = next_row, next_col

        # Increment the number
        num += 1

    return magic_square

def print_magic_square(magic_square):
    for row in magic_square:
        print(row)

# Create and print the magic square
magic_square = create_magic_square()
print("Magic Square:")
print_magic_square(magic_square)
```
